## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# import Dataset

```python
df = pd.read_csv('train (3).csv',usecols=['FireplaceQu','SalePrice','GarageQual'])
```

```python
df.sample(10)
```

```python
df.head()
```

# Mode - imputation

# Check missing Values

```python
df.isnull().mean()*100
```

# Plot Bar Garage Value

```python
df['GarageQual'].value_counts().plot(kind='bar')
```

```python
df['GarageQual'].mode()
```

# KDE PLOT | Compare Houses With TA

```python
fig = plt.figure()
ax = fig.add_subplot(111)
df[df['GarageQual']=='TA'].plot(kind='kde',ax=ax)
df[df['GarageQual'].isnull()]['SalePrice'].plot(kind='kde',ax=ax,color='red')
lines, lables = ax.get_legend_handles_labels()
lables = ['Houses with TA', 'Houses with NA']
ax.legend(lines, lables, loc='best')
plt.title('GarageQual')
```

# Store variable TA in temp

```python
temp =df[df['GarageQual'] == 'TA'] ['SalePrice']
```

# Replace missing value with TA

```python
df['GarageQual'].fillna('TA',inplace=True)
```

#  Review Bar Plot Changes

```python
df['GarageQual'].value_counts().plot(kind='bar')
```

# Draw plot again |  after Imputation

```python

fig = plt.figure()
ax = fig.add_subplot(111)
temp.plot(kind='kde',ax=ax)
df[df['GarageQual'] == 'TA']['SalePrice'].plot(kind='kde',ax=ax,color='red')
lines, lables = ax.get_legend_handles_labels()
lables = ['Original variable', 'Imputed variable']
ax.legend(lines, lables, loc='best')
plt.title('GarageQual')
```
